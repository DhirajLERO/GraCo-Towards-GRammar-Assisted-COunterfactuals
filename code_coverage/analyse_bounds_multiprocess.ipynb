```python
def load_data(data):
    return np.load(data)


def apply_phenotype(array, phenotype):
    # Split the phenotype string into individual operations
    operations = phenotype.split(';')
    x = copy.deepcopy(array)
    # Iterate over each operation and execute it
    for operation in operations:
        # Strip any leading/trailing whitespace from the operation
        operation = operation.strip()
        
        # Use the exec function to execute the operation on the array
        if operation:
            exec(operation)

    return x


def check_bounds(np_array, bounds):
 
    # Ensure the input is a NumPy array
    np_array = np.asarray(np_array)

    # Check if the length of bounds matches the length of the array
    if len(np_array) != len(bounds):
        raise ValueError("Length of bounds must match the length of the numpy array")

    # Iterate over each element and its corresponding bounds
    for idx, (value, (min_bound, max_bound)) in enumerate(zip(np_array, bounds)):
        if not (min_bound <= value <= max_bound):
            return False

    return True

def check_bounds_on_all_phenotype(phenotype_list, o_1_list,  bounds, input_data):
    for i, phenotype in enumerate(phenotype_list):
        evolved_x = apply_phenotype(input_data, phenotype)
        if check_bounds(evolved_x, bounds):
            print("within bounds")
            print("the value of fitness: ", o_1_list[i])
        else:
            print("out of bounds")
            print("the value of fitness: ", o_1_list[i])

bounds = [(18, 80), (0, 1), (0, 3), (0, 2), (0, 3), (0, 2), (100.0, 50000.0), (3, 100), (0, 7)]
```

```python
input_data = load_data("output/NSGAIII_test/0/input_data.npy")
phenotype_list = data['Phenotype'].to_list()
o_1_list = data['o_1'].to_list()


check_bounds_on_all_phenotype(phenotype_list, o_1_list,  bounds, input_data)

        
```
